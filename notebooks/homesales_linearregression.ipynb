```python
conv_BsmtQual = [4 if values == 'Ex' else 3 if values == 'Gd' else 2 if values == 'TA' else 1 if values == 'Fa' else 0 for values in homeprice_train.BsmtQual]
homeprice_train['BsmtQual'] = conv_BsmtQual
plt.scatter(homeprice_train.BsmtQual,homeprice_train.SalePrice)
plt.title("Price VS BsmtQual")
```

```python
conv_Street = [1 if values == 'Pave' else 0 for values in homeprice_train.Street]
homeprice_train['Street'] = conv_Street
plt.scatter(homeprice_train.Street,homeprice_train.SalePrice)
plt.title("Price VS Street")
```

```python
plt.scatter(homeprice_train.OverallCond,homeprice_train.SalePrice)
plt.title("Price VS OverallCond")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
labels = homeprice_train['SalePrice']
train1 = homeprice_train[['LotArea','OverallQual','OverallCond','BsmtFinSF1','TotalBsmtSF','2ndFlrSF','TotRmsAbvGrd','GarageCars','WoodDeckSF']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train1, labels, test_size = 0.6, random_state = 1234)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```
