```python
import pandas as pd
import numpy as np
import subprocess
import sys
import tkinter as tk
from tkinter import filedialog
```

```python
e1 = pd.read_excel('/workspaces/PracticeMachineLearning/data/ready aging.xlsx')
e2 = pd.read_excel('/workspaces/PracticeMachineLearning/data/AP APPROVALS - 07.12.23.xlsx')
```

```python
#replace all int values in the Vendor column with Nan
e1['Vendor'] = e1['Vendor'].mask(e1['Vendor'].apply(lambda x: isinstance(x, int)), np.nan)
```

```python
# Drop rows that has 'S' or 'T' oin the 'S' column
e1 = e1[~e1['S'].isin(['S', 'T'])]
```

```python
#copy the vendor code to each empty cell in row 'Vendor' if it is empty 
#with the cell above
for i in range(2, len(e1)):
    if pd.isnull(e1.at[i, 'Vendor']):
        e1.at[i, 'Vendor'] = e1.at[i-1, 'Vendor']
```

```python
# Drop rows based on missing the 'O' value. Invoices only show in 'O' rows
e1 = e1[e1['S'] == 'O']
```

```python
# rename e2 title column and drop first 15 characters
e2 = e2.rename(columns={'Title': 'invoiceid'})
e2['invoiceid'] = e2['invoiceid'].str[15:]
```

```python
# for e1 merge vendor and invoice to invoiceid
e1["invoiceid"] = e1['Vendor'].astype(str) +" "+ e1['Invoice'].astype(str)
```
